***Doing Some Data Preprocessing like Normalistaion,Scaling and Doing EDA***

```
def obese_classifier(X):
  if X<np.float64(18.5):
    return "U"
  elif (X>=np.float64(18.5)) & (X<=np.float64(24.9)):
    return "N"
  elif (X>np.float64(24.9)) & (X<=np.float64(29.9)):
    return "Ov"
  else:
    return "Ob"
```
